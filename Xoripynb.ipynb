```python
import numpy as np
import matplotlib.pyplot as plt

input_size = 2
hidden_size = 2
output_size = 1

learning_rate = 0.05
iterations = 50000
print_interval = 10000

X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y = np.array([
    [0],
    [1],
    [1],
    [0]
])

np.random.seed(42)   # randomness fixed

w1 = np.random.uniform(-1,1,(input_size, hidden_size))
w2 = np.random.uniform(-1,1,(hidden_size, output_size))

b1 = np.zeros((1, hidden_size))
b2 = np.zeros((1, output_size))

print("Initial Parameters:")
print("w1:\n", w1)
print("w2:\n", w2)
print("b1:", b1)
print("b2:", b2)


def sigmoid(z):
    return 1/(1 + np.exp(-np.clip(z,-250,250)))

def sigmoid_derivative(a):
    return a * (1 - a)


def forward(x):
    z1 = np.dot(x, w1) + b1
    a1 = sigmoid(z1)

    z2 = np.dot(a1, w2) + b2
    a2 = sigmoid(z2)

    return a1, a2


for i in range(iterations):

    a1, output = forward(X)

    error = y - output

    d_output = error * sigmoid_derivative(output)

    d_hidden = np.dot(d_output, w2.T) * sigmoid_derivative(a1)

    w2 += learning_rate * np.dot(a1.T, d_output)
    b2 += learning_rate * np.sum(d_output, axis=0, keepdims=True)

    w1 += learning_rate * np.dot(X.T, d_hidden)
    b1 += learning_rate * np.sum(d_hidden, axis=0, keepdims=True)

    if i % print_interval == 0:
        loss = np.mean(error ** 2)
        print(f"Iteration {i}, Loss: {loss:.6f}")


print("\nFinal Output:")
_, final_output = forward(X)
print(final_output)

print("\nXOR Result")
print("Input  Output")
for i in range(len(X)):
    print(X[i], round(final_output[i][0]))
```
